```python
from gtts import gTTS
# import pyttsx3 # Bibliothèque de synthèse vocale
import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup
from audioplayer import AudioPlayer
```

```python
def convert_epub_to_text(epub_path):
    # Ouvre le fichier EPUB
    try:
        book = epub.read_epub(epub_path)
    except Exception as e:
        print(f"Erreur lors de l'ouverture du fichier EPUB : {e}")
        return None

    full_text = ""

    # Définir l'intervalle de chapitres que vous voulez extraire (par exemple, du 1er au 20e)
    # L'indexation des chapitres commence à 0
    # end_chapter termine a la fin du chapitre donc si
    # start_chapter = 0 et end_chapter = 1 ... les suivants seront
    # start_chapter = 2 et end_chapter = 3
    # same number for each chapter (start = 0 and end = 0)
    start_chapter = 0  # 1er chapitre
    end_chapter = 0   # 20e chapitre
    # Récupère tous les éléments de type 'document' (qui sont les chapitres)
    text_items = [item for item in book.get_items() if item.get_type() == ebooklib.ITEM_DOCUMENT]

    # S'assurer que l'intervalle est valide
    if end_chapter >= len(text_items):
        print(f"L'intervalle spécifié est trop grand. Le livre n'a que {len(text_items)} chapitres.")
        end_chapter = len(text_items) - 1

    # Parcourir les éléments de texte dans l'intervalle spécifié
    for i in range(start_chapter, end_chapter + 1):
        chapter = text_items[i]
        # Décode le contenu de la page en texte
        page_content = chapter.content.decode('utf-8')
        # Utilise BeautifulSoup pour extraire uniquement le texte
        soup = BeautifulSoup(page_content, 'html.parser')
        text = soup.get_text(separator=' ', strip=True) # Utilise un espace comme séparateur et nettoie les espaces inutiles

        full_text += text + " "

    # for chapter in text_items:
    #     # Décode le contenu HTML du chapitre
    #     chapter_content = chapter.content.decode('utf-8')
    #     # Utilise BeautifulSoup pour extraire uniquement le texte
    #     soup = BeautifulSoup(chapter_content, 'html.parser')
    #     text = soup.get_text(separator=' ', strip=True) # Utilise un espace comme séparateur et nettoie les espaces inutiles
    #     full_text += text + " "
    return full_text

# Chemin vers ton fichier EPUB
epub_file_path = 'book/impossible_retour.epub'
book_text = convert_epub_to_text(epub_file_path)

print(book_text)
```

## Use PYTTSX3

### (always restart the kernel because pyttsx engine still running in interactive environment like Jupyter Notebook)

```python
import os
import sys
import subprocess

# Le texte à convertir
# book_text = "allo"

# Chemins des fichiers
output_dir = 'mp3'
m4a_file_path = os.path.join(output_dir, 'temp_audio.m4a')
mp3_file_path = os.path.join(output_dir, 'test.mp3')

# Crée le dossier de sortie s'il n'existe pas.
try:
    os.makedirs(output_dir, exist_ok=True)
except Exception as e:
    print(f"Erreur: Impossible de créer le dossier {output_dir}. Vérifiez vos permissions.")
    sys.exit()

# 1. Génère un fichier M4A temporaire avec la commande 'say'.
# -v VOICE
# -r RATE
say_command = f'say -v Amélie -r 170 -o "{m4a_file_path}" --file-format=m4af "{book_text}"'
print("Génération du fichier M4A en cours...")
say_result = subprocess.run(say_command, shell=True, capture_output=True, text=True)

if say_result.returncode != 0:
    print(f"Erreur : La commande 'say' a échoué. Code de sortie : {say_result.returncode}")
    print(f"Détails : {say_result.stderr}")
    sys.exit()

# 2. Convertit le fichier M4A en MP3 avec FFmpeg.
ffmpeg_command = f'ffmpeg -i "{m4a_file_path}" -q:a 0 -map a "{mp3_file_path}" -y'
print("Conversion du fichier en MP3 en cours...")
ffmpeg_result = subprocess.run(ffmpeg_command, shell=True, capture_output=True, text=True)

if ffmpeg_result.returncode == 0:
    print(f"Le fichier audio MP3 a été sauvegardé sous : {mp3_file_path}")
    # Nettoie le fichier temporaire
    os.remove(m4a_file_path)
else:
    print(f"Erreur: La conversion avec FFmpeg a échoué. Code de sortie: {ffmpeg_result.returncode}")
    print(f"Détails : {ffmpeg_result.stderr}")
```
